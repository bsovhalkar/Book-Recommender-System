```python
ratings = pd.read_csv("BX-Book-Ratings.csv", sep=";", on_bad_lines='skip',encoding= "latin-1")
```

```python
ratings.head()
```

```python
ratings.rename(columns={"User-ID" : "user_id","ISBN" : 'isbn',"Book-Rating" : "rating"},inplace=True)
```

```python
ratings.head()
```

```python
x =ratings['user_id'].value_counts()>200
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings['user_id'].value_counts().shape
```

```python
ratings.head()
```

```python
ratings_with_books = ratings.merge(books,on='isbn') 
```

```python
ratings_with_books.head(5)
```

```python
ratings_with_books.shape
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating' : 'no of rating'},inplace=True)
```

```python
number_rating.head(5)
```

```python
final_rating = ratings_with_books.merge(number_rating,on='title')
final_rating.head()
```

```python
final_rating =final_rating[final_rating['no of rating']>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
final_rating.head()
```

```python
book_pivot = final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot.head()
```

```python
book_parse = csr_matrix(book_pivot)
```

```python
type(book_parse)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_parse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
recommended = []
for i in range(len(suggestions)):
    recommended.append(book_pivot.index[suggestions[i]])
```
